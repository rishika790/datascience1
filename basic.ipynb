```python
import numpy as np

import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5])

y = np.array([1, 5, 6, 8, 9])

c = np.array(['red', 'green', 'blue', 'yellow', 'black'])

plt.scatter(x, y, c=c, alpha=0.5)

plt.show()
```

```python
x = np.array([1,2,3,4])
plt.pie(x)
plt.show()
```

```python
x = np.array([1,2,3,4])

labels =["apple", "banana", "dates","work"]
plt.pie(x, labels=labels)
plt.show()
```

```python
x =np.array([1,2,3,4])
labels =["apple", "banana", "dates","work"]
myexplode = [1,2,3,4]
plt.pie(x, labels=labels, explode=myexplode)
plt.show()
```

```python
x =np.array([1,2,3,4])

labels =["apple", "banana", "dates","work"]
myexplode = [0,1,2,3] 
c =['red', 'green', 'yellow', 'black']
plt.pie(x, labels=labels, explode=myexplode, colors=c)
plt.show()
```

```python
x =np.array([1,2,3,4])
labels =["apple", "banana", "dates", "work"] 
myexplode = [0,1,2,3] 
c =['red', 'green', 'yellow', 'black'] 
plt.pie(x, labels=labels, explode=myexplode, colors=c, autopct='%1.1f%%')
plt.show()
```

```python
x =np.array([1,2,3,4])
myexplode = [0,1,2,3]
labels =["apple", "banana", "dates", "work"] 
c =['red', 'green', 'yellow', 'black'] 
plt.pie(x, labels=labels, explode=myexplode, colors=c, autopct='%1.1f%%')
plt.legend(title = "my pie chart")
plt.show()
```
